```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a Pandas DataFrame
dk = pd.read_csv(r"C:\Users\gauri\Downloads\archive (3)\Unemployment in India.csv")

# Create a histogram with hue
plt.title('Unemployment in India')
sns.histplot(x=" Estimated Labour Participation Rate (%)",hue="Region",data=dk)
plt.show()
```

```python
# Line graph
plt.plot(dk['Date'], dk['Estimated Unemployment Rate (%)'])
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate over Time')
plt.show()

# Bar graph
region_data = dk.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
region_data.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Unemployment Rate by Region')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets into separate Pandas DataFrames
dk1 = pd.read_csv(r"C:\Users\gauri\Downloads\archive (3)\Unemployment in India.csv")
dk2 = pd.read_csv(r"C:\Users\gauri\Downloads\archive (3)\Unemployment_Rate_upto_11_2020.csv")

# Extract the columns containing the data for the histograms
data1 = dk1[' Estimated Labour Participation Rate (%)']
data2 = dk2[' Estimated Labour Participation Rate (%)']

# Create a figure with subplots for the histograms
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Plot the histograms for File 1
axes[0].hist(data1, bins=10, edgecolor='black')
axes[0].set_xlabel(' Estimated Labour Participation Rate (%)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram - File 1')

# Plot the histograms for File 2
axes[1].hist(data2, bins=10, edgecolor='black')
axes[1].set_xlabel(' Estimated Labour Participation Rate (%)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram - File 2')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the histograms
plt.show()
```
